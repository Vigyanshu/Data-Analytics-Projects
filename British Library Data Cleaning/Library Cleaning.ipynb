```python
extract_date = data['Date of Publication'].str.extract(r'^(\d{4})', expand = False)
extract_date.head()
```

```python
university_towns = []
with open('D:/Data Analyst Projects/British Library Data Cleaning/university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            university_towns.append((state, line))
university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns, columns=['Sate', 'Region Name'])
towns_df.head()
```

```python
def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
```

```python
towns_df = towns_df.applymap(get_citystate)
towns_df.head(10)
```
